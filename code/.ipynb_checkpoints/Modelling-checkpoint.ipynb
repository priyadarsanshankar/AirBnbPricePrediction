### Support Functions

##### Define a function to Show Culumns with missing values and their count

````python
def show_cols_with_missing(df):
    missing_vals = df.isnull().sum().sort_values(ascending = False)
    
    return missing_vals.iloc[missing_vals.nonzero()[0]]
````

##### Define a function to build ROC curve for multiclass classification

````python
def plot_roc(y_actual,y_calc,modname, n_classes = 3):
    t1=sum(x==0 for x in y_calc-y_actual)/len(y_calc)
    fpr = dict()
    tpr = dict()
    roc_auc = dict()
    for i in range(n_classes):
        fpr[i], tpr[i], _ = roc_curve(np.array(pd.get_dummies(y_actual))[:, i], np.array(pd.get_dummies(y_calc))[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])
    all_fpr = np.unique(np.concatenate([fpr[i] for i in range(n_classes)]))
    mean_tpr = np.zeros_like(all_fpr)
    for i in range(n_classes):
        mean_tpr += interp(all_fpr, fpr[i], tpr[i])
    mean_tpr /= n_classes
    fpr["macro"] = all_fpr
    tpr["macro"] = mean_tpr
    roc_auc["macro"] = auc(fpr["macro"], tpr["macro"])
    lw=2
    plt.figure(figsize=(8,8))
    plt.plot(fpr["macro"], tpr["macro"],
    label='macro-average ROC curve (area = {0:0.2f})'
          ''.format(roc_auc["macro"]),
    color='green', linestyle=':', linewidth=4)
    colors = cycle(['aqua', 'darkorange', 'cornflowerblue'])
    for i, color in zip(range(n_classes), colors):
        plt.plot(fpr[i], tpr[i], color=color, lw=lw,
             label='ROC curve of class {0} (area = {1:0.2f})'
             ''.format(i, roc_auc[i]))
    plt.plot([0, 1], [0, 1], 'k--', lw=1)
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(str('Receiver Operating Characteristics for '+modname))
    plt.legend(loc="lower right")
    plt.show()
````

##### Relative Error

````python
def weighted_classification_err(con_mat): 
    err = []
    labs = []
    for i in range(len(con_mat)):
        err.append((con_mat[i,:].sum() - con_mat[i,i])/(con_mat[i,:].sum()))
        labs.append(i)
    return pd.DataFrame(list(zip(labs,err)),columns=['Labels','Relative classification Error'])
````

##### MAPE

````python
def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
````

##### SMAPE

````python
def symmetric_mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return 100/len(y_true) * np.sum(2 * np.abs(y_pred - y_true) / (np.abs(y_true) + np.abs(y_pred)))
````

````python
#Select columns and perform the test train split
train_cols = [
    'log_price',
    #'property_type',
    #'room_type',
    #'amenities',
    'accommodates',
    #'bed_type',
    #'cancellation_policy',
    'cleaning_fee',
    #'city',
    'host_has_profile_pic',
    'host_identity_verified',
    #'host_response_rate',
    #'host_since',
    'instant_bookable',
    #'latitude',
    #'longitude',
    #'name',
    #'Neighbourhood',
    #'number_of_reviews',
    #'review_scores_rating',
    #'thumbnail_url',
    'dist_to_attr1',
    'dist_to_attr2',
    'dist_to_attr3',
    'dist_to_station',
    #'year_of_listing',
    #'month_of_listing',
    'days_since_hosted',
    #'bathrooms',
    #   'bedrooms',
    'bedroom_share',
    'bathroom_share',
    'room_type_Entire home/apt',
    'room_type_Private room',
    'room_type_Shared room',
    'city_Boston',
    'city_Chicago',
    'city_DC',
    'city_LA',
    'city_NYC',
    'city_SF',
    'description_length',
    'sentiment_intensity',
    # 'lda_attractions',
    # 'lda_transport',
    # 'lda_amenities',
    # 'lda_property_desc',
    '24-hour check-in', 
    'Bathtub',
    'Breakfast', 
    'Cooking Allowed', 
    'Elevator', 
    'Essentials',
    'Family/kid friendly', 
    'Fire and Safety Features', 
    'Free parking',
    'Gym', 
    'Hair dryer', 
    'Hangers', 
    'Internet', 
    'Iron',
    'Laptop friendly workspace', 
    'Laundry Facility', 
    'Lockbox',
    'Pet Friendly', 
    'Pool', 
    'Privacy Centric',
    'Security', 
    'Smoking allowed', 
    'Suitable for events', 
    'TV',
    'property_type_Apartment',
    'property_type_Condominium',
    'property_type_House',
    'property_type_Loft',
    'property_type_Townhouse',
    'cancellation_policy_flexible',
    'cancellation_policy_moderate',
    'cancellation_policy_strict'
]
              
target_col = 'rating_bin_ep'

np.random.seed(2018)
train = np.random.choice([True, False], listings_classification.shape[0], replace=True, p=[0.8, 0.2])
listings_lgbm_rating_train = listings_classification.iloc[train,:]
listings_lgbm_rating_test = listings_classification.iloc[~train,:]

x_train = listings_lgbm_rating_train[train_cols].values
x_test = listings_lgbm_rating_test[train_cols].values
y_train = listings_lgbm_rating_train[target_col].values
y_test = listings_lgbm_rating_test[target_col].values
y_train = y_train.ravel()
y_test = y_test.ravel()

print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
````

````python
#Create dataset for lightgbm
lgb_train = lgb.Dataset(x_train, y_train)
lgb_eval = lgb.Dataset(x_test, y_test, reference=lgb_train)

#Config the LGBM model parameters
params = {'task': 'train',
    'boosting_type': 'gbdt',
    'objective': 'multiclass',
    'num_class':3,
    'metric': 'multi_logloss',
    'learning_rate': 0.05,
    'max_depth': 7,
    'num_leaves': 17,
    'feature_fraction': 0.4,
    'bagging_fraction': 0.6,
    'bagging_freq': 17
         }


print('Starting training...')
# train

evals_result = {}

gbm = lgb.train(params,
                lgb_train,
                num_boost_round=1000,
                valid_sets=lgb_eval,
                early_stopping_rounds=5,
               evals_result=evals_result,
               feature_name=train_cols)

print('Saving model...')
#Save the fit model to a file
gbm.save_model('model.txt')
````

````python
print('Starting predicting...\n')

#Predict
y_pred = gbm.predict(x_test, num_iteration=gbm.best_iteration)
predictions = []
for x in y_pred:
    predictions.append(np.argmax(x))

print('Test scores \n')
#Test dataset scores
print('Precision:',round(precision_score(y_test,predictions,average='weighted'),3))
print('Recall:',round(recall_score(y_test,predictions,average='weighted'),3))
print('F1:',round(f1_score(y_test,predictions,average='weighted'),3))
print('Accuracy:',round(accuracy_score(y_test,predictions),3))

#Predict
y_pred = gbm.predict(x_train, num_iteration=gbm.best_iteration)
predictions = []
for x in y_pred:
    predictions.append(np.argmax(x))
print('\nTrain scores \n')
#Train dataset scores
print('Precision:',round(precision_score(y_train,predictions,average='weighted'),3))
print('Recall:',round(recall_score(y_train,predictions,average='weighted'),3))
print('F1:',round(f1_score(y_train,predictions,average='weighted'),3))
print('Accuracy:',round(accuracy_score(y_train,predictions),3))
````
